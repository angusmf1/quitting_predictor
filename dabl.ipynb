```python
import pandas as pd
import dabl

data = pd.read_csv('new_combined_with_text_data.csv')
```

```python
types = dabl.detect_types(data)
print(types)
```

```python
dabl.clean(data)
```

```python
dabl.plot(data, target_col="Quitter")


















```

```python
columns = ['education level', 'title_1', 'company_1', "level", "total_exp", "exp_check", "Quitter", "custom_url",
           "Region", "Word_Count"]

''' 
           "job_desc_application", "job_desc_data", "job_desc_design", "job_desc_development", 
           "job_desc_project", "job_desc_service", "job_desc_software", "job_desc_system", "job_desc_team",
           "job_desc_using", "summary_application", "summary_data", "summary_development", "summary_experience", 
           "summary_software", "summary_system", "summary_team", "summary_technology", "summary_web", 
           "summary_year", "skills_agile_methodology", "skills_cs", "skills_html", "skills_java", 
           "skills_javascript", "skills_linux", "skills_mysql", "skills_python", "skills_software_development",
           "skills_sql", "Word_Count", "summary_is_null", "skills_is_null"]
'''

df = data[columns]
```

**Simple Classifier Model**

```python
ec = dabl.SimpleClassifier(random_state=0).fit(df, target_col='Quitter') 
```

```python
dabl.explain(ec)
```

**Enhanced Classifier Model**

```python
ac = dabl.AnyClassifier().fit(df, target_col='Quitter') 
```

```python
dabl.explain(ac)
```
