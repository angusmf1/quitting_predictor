```python
import pandas as pd
import dabl

data = pd.read_csv('ready_train_w_urlcheck.csv')
```

```python
types = dabl.detect_types(data)
print(types)
```

```python
dabl.clean(data)
```

```python
selected_columns = ['education level', 'title_1', 'company_1', 'level', 'total_exp', 'custom_url', 'State','Quitter']
df = data[selected_columns]
```

**Simple Classifier Model**

```python
ec = dabl.SimpleClassifier(random_state=0).fit(df, target_col='Quitter') 
```

```python
dabl.explain(ec)
```

**Enhanced Classifier Model**

```python
ac = dabl.AnyClassifier().fit(df, target_col='Quitter') 
```

```python
dabl.explain(ac)
```
